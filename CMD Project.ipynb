### Section 1: The Hipparcos CMD for The Local Stellar Neighbourhood

We will use the data from Hipparcos observation. The dataset is in the file 'Hipparcos.dat'. 

There are two header rows that we will need to skip when reading in the data. The columns we will use are:
* p: parallax in milliarcseconds
* sig_p: parallax uncertainty in milliarcseconds
* v: apparent V magnitude in mag
* bv: B-V color in mag
* sig_bv: B-V uncertainty in mag

Recall that astronomers use different filters to observe stars and estimate their colour (which is related to their effective temperatures). With the Hipparchos data, we are looking at the stellar brightness in the V (visual) and B (blue) filters. 

```python
# Import the needed libraries for calculations and plots

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Read in the data using np.loadtxt and create separate arrays for each column

p, sig_p, v, bv, sig_bv = np.loadtxt('Hipparcos.dat', usecols = (0, 1, 2, 3, 4), skiprows = 2, unpack = True)
```

Recall that a colour magnitude diagram (CMD) usually plots the absolute visual magnitude (V) against the colour. Right now we only have apparent magnitudes. This isn't a problem for colour; the B-V colour will remain the same if we calculate it using apparent or absolute magnitudes! As a first look at a CMD, we can plot the apparent visual vs. the B-V colour. The B-V colour is simply the difference in B and V magnitudes. 

```python
# Using a scatter plot, plot the CMD of the Hipparchos data as apparent V magnitude vs B-V colour
# Because there are so many stars, make the size of the plotted points smaller by setting the parameter s in the plt.scatter

fig = plt.figure()
plt.scatter(bv, v, color = 'darkorange', s = 1)

# Add title and axis labels

plt.title('Plot of Apparent V Magnitude vs (B - V) Colour from Hipparcos Catalogue')
plt.xlabel('(B - V)')
plt.ylabel('Apparent V Magnitude')

# Since lower magnitudes are brighter, we have to invert the axis

plt.gca().invert_yaxis()

# Save the figure

plt.savefig('Plot of Apparent Magnitude and Colour Index - Hipparcos.png')
```

As we can see, this CMD does not give a clear trend suggestive of a main sequence. Thus, to understand the trend of a main sequence, let's inspect the distribution of the parallax. To accomplish this, we will make a histogram of the parallax values in the dataset.

```python
# Make a histogram of the parallax values using plt.hist

plt.hist(p, bins = 'auto', range = (10, 30), color = 'darkorange')

plt.title('Histogram of The Parallax Values of Stars from Hipparcos Catalogue')
plt.xlabel('Parallax (milliarcseconds)')
plt.ylabel('Number of Stars')

plt.show()
```

As we can see, we have a wide range of parallax values. This suggests that the stars are all spread out across all directions and in various distances from the sun. Next, we are going to define a function that will give us the absolute magnitude of the stars from the given parallax and apparent magnitude.

```python
# Write a function to calculate absolute magnitude given the apparent magnitude and parallax

def get_absolute(apparent: list[float], parallax: list[float]) -> list[float]:
    """
    Return a list of absolute magnitudes from the given list of apparent magnitudes and the given list of parallax values.
    # Precondition: Length of apparent is the same as the length of parallax. Given parallax is in milliarcseconds.
    """
    list_of_absolute = []
    for i in range(len(apparent)):
        parallax_in_arcseconds = parallax[i] * 10 ** (-3)
        distance = 1 / parallax_in_arcseconds
        absolute = apparent[i] + 5 -(5 * math.log(distance, 10))
        list_of_absolute.append(round(absolute, 2))
    return list_of_absolute
```

```python
# Since our data is in arrays, to use this function, we have to first convert the array into list

list_of_apparent = v.tolist()
list_of_parallax = p.tolist()

# Calculate the absolute V magnitudes using the function

absolute_v = get_absolute(list_of_apparent, list_of_parallax)
```

Now we are going to replot the CMD, using the absolute V magnitudes and original B-V colour.

```python
# Plot the CMD

fig = plt.figure()

plt.scatter(bv, absolute_v, color = 'darkorange', s = 1)

plt.title('Plot of Absolute V Magnitude vs (B - V) Colour from Hipparcos Catalogue')
plt.xlabel('(B - V)')
plt.ylabel('Absolute V Magnitude')

plt.gca().invert_yaxis()

plt.show()
```

On this plot, we are going to add the sun. Appendix A in Carroll & Ostlie gives the Sun's absolute bolometric magnitude $M_B$, and the apparent blue and visual magnitudes. Add a red star to overlay the Sun's location on the plot above using a second `plt.scatter` command. Set the marker style in `plt.scatter` using `marker = '*'`, and change the colour using `c = 'red'`.  

```python
# The absolute magnitude and BV of sun from Carroll & Ostlie

M_sun = 4.74
BV_sun = 0.65

# Add sun on the CMD plot

fig = plt.figure()

plt.scatter(bv, absolute_v, color = 'darkorange', s = 1)
plt.plot(BV_sun, M_sun, marker = '*', color = 'red')
plt.text(BV_sun + 0.01, M_sun - 0.25, 'Sun', size = 12.5) 

plt.title('Plot of Absolute V Magnitude vs (B - V) Colour from Hipparcos Catalogue')
plt.xlabel('(B - V)')
plt.ylabel('Absolute V Magnitude')

plt.gca().invert_yaxis()

plt.savefig('Plot of Absolute Magnitude and Colour Index - Hipparcos.png')
```

### Section 2: Gaia Data for The Hyades Cluster

The Hipparchos data includes all stars within a particular distance of the Sun. These stars were born at different times, in different locations, and therefore we see stars in a range of evolutionary stages in the plot. Some massive stars are still on the main sequence! Some less massive stars have evolved into giants.

Next, we will investigate the data for a single stellar cluster, where we can assume that all stars formed at roughly the same time. Hopefully, this means that we can see clearly the main sequence, and the main sequence turn-off where the most massive stars have already exhausted their hydrogen in their cores, and have started evolving onto the giant branch. 

We will use data from the Gaia archive. The Gaia mission is measuring the magnitudes, locations, parallaxes and proper motions of stars within 10,000 pc of the Sun. The Hyades data are in the file 'Gaia_Hyades.csv'. This is the output from a search of the Gaia database [https://gea.esac.esa.int/archive/](here). The Gaia archive contains a lot of data on a lot of stars! So, to ease the analysis, the dataset has been limited to an area of 240 arcminutes around the centre of the Hyades cluster and contain only stars with parallaxes close to the expected value for the cluster. 

Here, we will first use the measured parallax and proper motion values to select the set of stars that are truly members of the Hyades cluster. Then we will convert apparent to absolute magnitudes, plot the CMD, and compare the CMD with several isochrones. Isochrones (same time) are tables that plot the expected CMD for a population of stars with a range of masses at a given age.

The Gaia data file contains the following columns:

* ra: right ascension in degrees
* dec: declination in degrees
* parallax: parallax in milliarcseconds
* pmra: proper motion in milliarcseconds per year, in right ascension
* pmdec: proper motion in milliarcseconds per year, in declination
* phot_g_mean_mag: apparent G magnitude in mag
* phot_bp_mean_mag: apparent B magnitude in mag
* phot_rp_mean_mag: apparent R magnitude in mag

In the data, some stars don't have data for all columns. Unfortunately this messes up `np.loadtxt`. To read these data in, use `np.genfromtxt()` with `delimiter=','`, and `skip_header` rather than `skiprows` (as in `np.loadtxt`). Remember to set `unpack=True`!

```python
# Read in Gaia data

ra, dec, parallax, parallax_error, pmra, pmdec, g_mag, b_mag, r_mag = np.genfromtxt('Gaia_Hyades.csv', delimiter = ',', skip_header = 1, unpack = True)
```

```python
# First, let's see where all the stars are located

fig = plt.figure()

plt.scatter(ra, dec, s = 1, color = 'darkorange')

plt.title('Plot of The Position of Stars around The Hyades Cluster')
plt.xlabel('Right Ascension (degrees)')
plt.ylabel('Declination (degrees)')

plt.show()
```

The data has already been limited by setting a minimum parallax of 15 milliarcseconds. Next, plot a histogram of the parallax values to see whether we need to limit the data further. A cluster of stars should all be at nearly the same parallax.

```python
# Look at parallax measurements as a histogram

plt.hist(parallax, bins = 'auto', range = (19, 23), color = 'darkorange')

plt.title('Histogram of The Parallax Values of Stars in The Hyades Cluster')
plt.xlabel('Parallax (milliarcseconds)')
plt.ylabel('Number of Stars')

plt.savefig('Histogram of Parallax - Hyades.png')
```

It is important to keep a note of what range of parallax values best identifies where the parallax peak contains the most stars, which in this case is from 19 to 23 milliarcseconds. 

Next, we will look at the proper motions of the stars. The proper motion is the observed motion of the star on the sky, once the effect of parallax is removed. Because Gaia is observing the sky over many years, with high precision, we are able to measure very small motions of the nearby stars relative to more distant sources. 

A cluster of stars, that formed from the same gas cloud, should also be moving at similar speeds and in similar directions. The values won't be exactly the same, as random motions in the original gas cloud result in some randomness in the stellar motions. But if we plot the proper motions of the stars in the catalog, hopefully we will see a clustering of stars at a particular velocity. It is also important to keep a note of what range of proper motions best identifies the stars in the cluster.

```python
# First, plot pmra vs pmdec for all the stars as a scatter plot 
# Zoom in by setting plt.xlim() and plt.ylim()

fig = plt.figure()

plt.scatter(pmdec, pmra, s = 1, color = 'darkorange')

plt.title('Plot of The Proper Motion of Stars around The Hyades Cluster')
plt.xlabel('Proper Motion in Declination (milliarcseconds per year)')
plt.ylabel('Proper Motion in Right Ascension (milliarcseconds per year)')

plt.xlim(-100, 50)
plt.ylim(-100, 200)

plt.savefig('Plot of Proper Motion - Hyades.png')
```

Next, we will use the limits in parallax and proper motion to select out only those stars that we feel are in the cluster. We do this by identifying the indices of stars whose proper motions and parallaxes fall within the identified ranges above. 

Then we can use these indices to select the values from the parameter lists of only those stars in the cluster for the following analysis. 

```python
# Here, fill in the best estimates for the minimum and maximum values of pmra and pmdec that best select the cluster stars only

pmra_min = 80
pmra_max = 120
pmdec_min = -45
pmdec_max = -10

# Also list the minimum and maximum parallax limits from the histogram:

par_min = 19
par_max = 23
```

```python
# Use the np.where() function to get a list of indices

cluster_indices = np.where((parallax > par_min) & (parallax < par_max) & (pmra > pmra_min) & (pmra < pmra_max) & (pmdec > pmdec_min) & (pmdec < pmdec_max))
```

```python
# Next, plot the star locations again in RA and Dec, but this time overlay the plot with the cluster stars in a different colour

fig = plt.figure()

plt.scatter(ra, dec, s = 0.5, color = 'darkorange')
plt.scatter(ra[cluster_indices], dec[cluster_indices], s = 1, color = 'black', label = 'Stars in Hyades Cluster')

plt.title('Plot of The Position of Stars around The Hyades Cluster')
plt.xlabel('Right Ascension (degrees)')
plt.ylabel('Declination (degrees)')

plt.legend()

plt.savefig('Plot of Position - Hyades.png')
```

```python
# Creating new lists of parameters that only the stars in the cluster have

ra_sorted = ra[cluster_indices]
dec_sorted = dec[cluster_indices]
parallax_sorted = parallax[cluster_indices]
parallax_error_sorted = parallax_error[cluster_indices]
pmra_sorted = pmra[cluster_indices]
pmdec_sorted = pmdec[cluster_indices]
g_mag_sorted = g_mag[cluster_indices]
b_mag_sorted = b_mag[cluster_indices]
r_mag_sorted = r_mag[cluster_indices]
```

Calculate the mean and standard deviation of the parallaxes. 

Use these values to calculate the distance to the Hyades cluster. To calculate the uncertainty in the result, use the standard deviation of the parallax values as the uncertainty in the parallax, and propagate that uncertainty in the distance calculation.

```python
# Calculate mean and standard deviation of the parallax values. 

parallax_mean = round(np.mean(parallax_sorted), 3)
parallax_stdev = round(np.std(parallax_sorted), 3)
parallax_uncertainty = round(np.mean(parallax_error_sorted), 3)

print("The mean of the parallax values is {}. \nThe standard deviation of the parallax values is {}. \nThe parallax measurement \
uncertainty is {}.".format(parallax_mean, parallax_stdev, parallax_uncertainty))
```

As we can see, the standard deviation is greater than the parallax measurement uncertainties. This means that each measurement of the parallax value is highly accurate, but the distribution of the parallax values of all the stars within the cluster is quite wide. Since the standard deviation is greater than the measurement uncertainties, the standard deviation will contribute more to the uncertainty of the distance measurement.

```python
# Calculate the mean cluster distance based on our mean parallax value. 
# Remember that these are in units of milliarcseconds.
# Calculate the uncertainty in the distance measurement, given the 
# standard deviation in the parallax values.

mean_distance = round((1 / (parallax_mean * (10**(-3)))), 3)
uncertainty_distance = round(((mean_distance * (parallax_stdev) / parallax_mean)), 3)

print("The mean distance of the cluster is {} pc with uncertainty of {} pc.".format(mean_distance, uncertainty_distance))
```

From Wikipedia (https://en.wikipedia.org/wiki/Hyades_(star_cluster)), the distance of the Hyades Cluster is around 47 pc which is similar to what we get here. Meanwhile, from the website of NASA (https://ntrs.nasa.gov/citations/19940024498), the distance of the Hyades Cluster is around 45 pc which is within our range of uncertainty.

In order to create the CMD, we need information about the absolute magnitude. Since each star has a parallax measurement, we won't use the mean value to convert from apparent magnitudes to absolute magnitudes. But, we will use the function `get_absolute` to calculate the absolute G magnitudes for the Hyades stars. 

```python
# Use the function get_absolute to find the absolute magnitude of the stars in the cluster. Since the function only accepts list,
# we have to first convert our arrays into lists

g_list = g_mag_sorted.tolist()
parallax_list = parallax_sorted.tolist()
G_mag = get_absolute(g_list, parallax_list)
```

Next plot the Hyades CMD. We need to calculate the B-R colour for the x axis (rather than B-V). Overlay the Sun again!

Assume B-R = 0.82 for the Sun ([Casagrande & VandenBerg 2018](https://academic-oup-com.myaccess.library.utoronto.ca/mnrasl/article/479/1/L102/5035821#118661072)).

```python
# Calculate B-R colour for the Hyades:

br = b_mag_sorted - r_mag_sorted

# Sun's B-R and G values:

G_sun = 4.67
br_sun = 0.82

# Plot the Hyades CMD

fig = plt.figure()

plt.scatter(br, G_mag, color = 'darkorange', s = 1)
plt.plot(br_sun, G_sun, marker = '*', color = 'red')
plt.text(br_sun, G_sun, 'Sun') 
plt.axhline(y = 1, color = 'red', linestyle = '--', label = 'Turn-off Line') 

plt.title('Colour - Magnitude Diagram of The Hyades Cluster')
plt.xlabel('(B - R)')
plt.ylabel('Absolute G Magnitude')

plt.gca().invert_yaxis()
plt.legend()

plt.show()
```

From the graph above, the main sequence turn off happens at an absolute magnitude of around 1. We can get the luminosity of the star by comparing it to the sun using the formula:

$M_{star} - M_{sun} = -2.5\log\frac{L_{star}}{L_{sun}}$ 

where $M$ is the absolute magnitude and $L$ is the luminosity
From C&O, $M_{sun} = 4.83$ and $L_{sun} = 3.8 \times 10^{26}$ watts. From the graph, we estimate $M_{star} = 1$

Thus, plugging in all the numbers, we get:

\begin{align*} 
1 - 4.83 = -2.5\log\frac{L_{star}}{L_{sun}}\\
\frac{-3.83}{-2.5} = \log\frac{L_{star}}{L_{sun}}\\
1.532 = \log\frac{L_{star}}{L_{sun}}\\
\frac{L_{star}}{L_{sun}} = 10^{1.532} = 34\\
L_{star} = 34 L_{sun}
\end{align*}

Using the relation of $L \sim M^4$, we can compare the star luminosity and the sun luminosity

\begin{align*}
\frac{L_{star}}{L_{sun}} = \left(\dfrac{M_{star}}{M_{sun}}\right)^4\\
\left(\dfrac{L_{star}}{L_{sun}}\right)^\frac{1}{4} = \frac{M_{star}}{M_{sun}}\\
34^\frac{1}{4} = \frac{M_{star}}{M_{sun}}\\
2.4 = \frac{M_{star}}{M_{sun}}\\
M_{star} = 2.4 M_{sun}
\end{align*}

Given the position of the Sun on the plot, it is reasonable for a star along the turn-off line to have a mass of around 2.4 solar masses.

Next, we calculate the nuclear timescale for a star with $2.4 M_{sun}$

The nuclear timescale is defined as the time it takes for a star to burn all its fuel under a constant burn rate. From C&O, 
it can be mathematically written as $t = \frac{0.1 \times 0.007 \times M_{star} \times c^2}{L_{star}}$ where $t$ is the nuclear timescale, $c$ is the speed of light with the value of $3 \times 10^8$ m/s, $M$ is the mass, and $L$ is the luminosity.

Plug in $M_{star} = 2.4 M_{sun}$ and $L_{star} = 34 L_{sun}$
\begin{align*}
t = \frac{0.1 \times 0.007 \times 2.4 M_{sun} \times c^2}{34 L_{sun}}
\end{align*}
From C&O, $M_{sun} = 1.99 \times 10^{30}$ kg and $L_{sun} = 3.8 \times 10^{26}$ watts.

Plugging in those values, we would get

\begin{align*}
t = \frac{0.1 \times 0.007 \times 2.4 \times 1.99 \times 10^{30} \times (3 \times 10^8)^2}{34 \times 3.8 \times 10^{26}} = 2.3 \times 10^{16} \text{s}\\
t = 2.3 \times 10^{16} \text{s} \times \left(\dfrac{1}{86400}\right) \text{day} \times \left(\dfrac{1}{365.25}\right) \text{year} \times \left(\dfrac{1}{10^6}\right) \text{Myr} = 738 \text{Myr}
\end{align*}

Thus, the nuclear timescale for a star with this mass is 738 Myr.

Next we will compare the CMD for the Hyades with a set of isochrones. These are tables of the predicted absolute magnitude and colour of a cluster of stars at a particular age.

For the Hyades, we want to use the files that include 'z02' in their filename - these are made for stars with solar metallicity (same fraction of metals to hydrogen as the Sun). 

An example filename is 'isochrone_z02_316Myr.dat'. 

The columns are:

* Mass: stellar mass in units of $M_\odot$
* Gmag: absolute G magnitude
* G_BPmag: absolute B magnitude (in the Gaia system)
* G_RPmag: absolute R magnitude (in the Gaia system)

```python
# Read in the isochrones

mass1G, Gmag1G, G_BPmag1G, G_RPmag1G = np.genfromtxt('isochrone_z02_1Gyr.dat', delimiter = ',', skip_header = 0, unpack = True)
mass1M, Gmag1M, G_BPmag1M, G_RPmag1M = np.genfromtxt('isochrone_z02_1Myr.dat', delimiter = ',', skip_header = 0, unpack = True)
mass316, Gmag316, G_BPmag316, G_RPmag316 = np.genfromtxt('isochrone_z02_316Myr.dat', delimiter = ',', skip_header = 0, unpack = True)
mass500, Gmag500, G_BPmag500, G_RPmag500 = np.genfromtxt('isochrone_z02_500Myr.dat', delimiter = ',', skip_header = 0, unpack = True)
mass750, Gmag750, G_BPmag750, G_RPmag750 = np.genfromtxt('isochrone_z02_750Myr.dat', delimiter = ',', skip_header = 0, unpack = True)
```

As we can see from the plot above, there are several differences in the isochrone shapes:
* The main sequence of an older isochrone is shorter than the main sequence of a younger isochrone.
* The main sequence turn off point for an older isochrone happens at a higher magnitude and higher index colour relative to the main sequence turn off point for a younger isochrone.The turn off point for an older isochrone is also much curvier than the main sequence turn off point for a younger isochrone.
* The giant branch for an older isochrone happens at a higher magnitude and higher colour index.
* Overall, the stars in an older cluster have evolved much further than the stars in a younger cluster because there is not many stars left in the main sequence.

From the graph above, the main sequence turn off happens at an absolute magnitude of around 0. We can get the luminosity of the star by comparing it to the sun using the formula:

$M_{star} - M_{sun} = -2.5\log\frac{L_{star}}{L_{sun}}$ 

where $M$ is the absolute magnitude and $L$ is the luminosity
From C&O, $M_{sun} = 4.83$ and $L_{sun} = 3.8 \times 10^{26}$ watts. From the graph, we estimate $M_{star} = 0$

Thus, plugging in all the numbers, we get:

\begin{align*} 
0 - 4.83 = -2.5\log\frac{L_{star}}{L_{sun}}\\
\frac{-4.83}{-2.5} = \log\frac{L_{star}}{L_{sun}}\\
1.932 = \log\frac{L_{star}}{L_{sun}}\\
\frac{L_{star}}{L_{sun}} = 10^{1.932} = 85.5\\
L_{star} = 85.5 L_{sun}
\end{align*}

Using the relation of $L \sim M^4$, we can compare the star luminosity and the sun luminosity

\begin{align*}
\frac{L_{star}}{L_{sun}} = \left(\dfrac{M_{star}}{M_{sun}}\right)^4\\
\left(\dfrac{L_{star}}{L_{sun}}\right)^\frac{1}{4} = \frac{M_{star}}{M_{sun}}\\
85.5^\frac{1}{4} = \frac{M_{star}}{M_{sun}}\\
3 = \frac{M_{star}}{M_{sun}}\\
M_{star} = 3 M_{sun}
\end{align*}

Given the position of the Sun on the plot, it is reasonable for a star along the turn-off line to have a mass of around 3 solar masses.

Next, we calculate the nuclear timescale for a star with $3 M_{sun}$

The nuclear timescale is defined as the time it takes for a star to burn all its fuel under a constant burn rate. From C&O, 
it can be mathematically written as $t = \frac{0.1 \times 0.007 \times M_{star} \times c^2}{L_{star}}$ where $t$ is the nuclear timescale, $c$ is the speed of light with the value of $3 \times 10^8$ m/s, $M$ is the mass, and $L$ is the luminosity.

Plug in $M_{star} = 3 M_{sun}$ and $L_{star} = 85.5 L_{sun}$
\begin{align*}
t = \frac{0.1 \times 0.007 \times 3 M_{sun} \times c^2}{85.5 L_{sun}}
\end{align*}
From C&O, $M_{sun} = 1.99 \times 10^{30}$ kg and $L_{sun} = 3.8 \times 10^{26}$ watts.

Plugging in those values, we would get

\begin{align*}
t = \frac{0.1 \times 0.007 \times 3 \times 1.99 \times 10^{30} \times (3 \times 10^8)^2}{85.5 \times 3.8 \times 10^{26}} = 1.1 \times 10^{16} \text{s}\\
t = 1.1 \times 10^{16} \text{s} \times \left(\dfrac{1}{86400}\right) \text{day} \times \left(\dfrac{1}{365.25}\right) \text{year} \times \left(\dfrac{1}{10^6}\right) \text{Myr} = 367 \text{Myr}
\end{align*}

Thus, the nuclear timescale for a star with this mass is 367 Myr.

As we can see from the plot, there are some differences between the plot of a high metallicity isochrone and the plot of a low metallicity isochrone.
* Notice that the main sequence turn-off point for a low metallicity star happens at a higher color index and at a higher magnitude compared to the main sequence turn-off point for a high metallicity star. It means that low metallicity stars evolve off the main sequence earlier before the the high metallicity stars. Therefore, we can conclude that low metallicity stars evolve faster than the high metallicity stars.
* Notice that after the turn-off point, the plot for the low metallicity stars goes a little bit upward for some time before steeply increasing to a very lumnious region. Meanwhile, for the high metallicity stars, the plot goes to the right quite steadily before gradually increasing to the luminous region. For the low metallicity stars, this process happens in a short amount of time compared to the high metallicity stars as the blue line is longer than the red line. This suggests that the high metallicity stars spend quite some time in this process.
